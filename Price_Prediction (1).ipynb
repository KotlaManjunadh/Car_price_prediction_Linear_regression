```python
# DATA PRE-PROCESSING 

# creating a deep copy of original dataframe
df1 = df.copy(deep = True)
```

```python
df1.head(6)
```

```python
# defining a function for label Encoding 
def l_encode(a):
  label_encoder = LabelEncoder()
  df1[a]= label_encoder.fit_transform(df1[a])
 

l_encode('CarName')
df1
```

```python
column_names = ["fueltype", 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation','enginetype', 'cylindernumber','fuelsystem']

for i in column_names :
  l_encode(i)
```

```python
df1
```

```python
scaler = StandardScaler()
df2 = scaler.fit_transform(df1)
df3 = pd.DataFrame(data= df2 , columns= df1.columns)
df3
```

```python
df3.corr()
```

```python
# splitting independent and dependant variables
y = df3['price']
x = df3.drop(columns= ['price','car_ID']  , axis =1)
```

```python
x
```

```python
y
```

```python
# checking correlation between variables

plt.figure(figsize=(20,20))
sb.heatmap(x.corr(), annot = True)
plt.style.use("dark_background")
plt.show()
```

```python
# we can see that there is a multicollinearity in this data
x2 = x.corr()

for i in range(x2.shape[0]):
  for j in range(i, x2.shape[1]):
    if ((abs(x2.iloc[i,j])>0.8)&(abs(x2.iloc[i,j])!=1)):
      print (x2.iloc[i,j], x2.columns[j],'&',x2.index[i] )
```

```python
# removing fueltype, carlength , citympg , curbweight ,enginesize to remove multicolinearity in data
x.drop(columns=['fueltype','carlength','citympg','curbweight','enginesize'], axis =1 , inplace=True)

x
```

```python
# train test split
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.25 ,random_state=9)

print(x_train.shape, x_test.shape , y_train.shape , y_test.shape)
```

```python
# fitting in model
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```
